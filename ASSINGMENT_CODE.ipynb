```python
import requests
```

```python
population_url = "https://raw.githubusercontent.com/NaftalyMunene/KENYAN_POPULATION/refs/heads/main/POPULATION_DATA.json"
```

```python
population_data = requests.get(population_url)
```

```python
type(population_data)
```

```python
population_data_json = population_data.json()
```

```python
type(population_data_json)
```

```python
import pandas 
```

```python
kenya_population_data = pandas.DataFrame(population_data_json)
```

```python
kenya_population_data.to_csv("kenya_population_data.csv")
```
